```python
import pandas as pd

# Load the dataset
data_w_genres = pd.read_csv('/Users/taief/Desktop/MusicTry/data_w_genres.csv') 

# Display the first few rows of the DataFrame to understand its structure
data_w_genres.head(20)
```

```python
import ast  # For safely evaluating a string as a list

data_w_genres['genres'] = data_w_genres['genres'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Explode the 'genres' column into multiple rows
exploded_data = data_w_genres.explode('genres')

# Verify the transformation
exploded_data.head(30)
```

```python
# Fill NaN in 'genres' with a placeholder value like 'Unknown'
exploded_data['genres'] = exploded_data['genres'].fillna('Unknown')

exploded_data.head(30)
```

```python
# Check for NaN values in each column
nan_counts = exploded_data.isnull().sum()
print(nan_counts)

# To get a more detailed view, you can check which rows have NaN values
nan_rows = exploded_data[exploded_data.isnull().any(axis=1)]
print(nan_rows)
```

```python
# save the exploded data to a new csv file
exploded_data.to_csv('/Users/taief/Desktop/MusicTry/exploded_data.csv', index=False)
```
